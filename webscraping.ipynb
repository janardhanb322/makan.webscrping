```python
import requests
```

```python
import bs4
```

```python
url="https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city"
```

```python
data=requests.get(url)
data
```

```python
type(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(data.text,"html")
```

```python
soup.find('div',attrs={"class":"title-line"}).text
```

```python
soup.find('span',attrs={"itemprop":"addressLocality"}).text
```

```python
soup.find('div',attrs={"data-type":"price-link"}).text
```

```python
soup.find('td',attrs={"class":"lbl rate"}).text
```

```python
soup.find('td',attrs={"class":"size"}).text
```

```python
soup.find('td',attrs={"class":"val"}).text
```

```python
prop=soup.find_all('div',attrs={"class":"title-line"})
```

```python
prop_name=[]
for i in prop:
    prop_name.append(i.text)
```

```python
print(len(prop_name))
```

```python
prop=[i.text for i in prop]
print(prop)
```

```python
a=soup.find_all('span',attrs={"itemprop":"addressLocality"})
location=[i.text for i in a]
print(location)
print(len(location))
```

```python
b=soup.find_all('div',attrs={"data-type":"price-link"})
link=[i.text for i in b]
print(link)
print(len(link))
```

```python
c=soup.find_all('td',attrs={"class":"lbl rate"})
rate=[i.text for i in c]
print(rate)
print(len(rate))
```

```python
d=soup.find_all('td',attrs={"class":"size"})
size=[i.text for i in d]
print(size)
print(len(size))
```

```python
e=soup.find_all('td',attrs={"class":"val"})
status=[i.text for i in e]
print(status)
print(len(status))
```

```python
import pandas as pd
```

```python
d=pd.DataFrame({"prop":prop,"location":location,"link":link,"rate":rate,"size":size,"status":status})
d
```

```python
d.to_csv("Hyd_Apartments All.csv")
```
